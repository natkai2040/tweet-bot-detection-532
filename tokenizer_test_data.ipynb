```python
import json
import pandas as pd
import numpy as np
```

```python
def get_file(filename):
    data = {}
    try:
        with open(filename, 'r') as f:
            data = json.load(f)
        print("JSON data loaded successfully:")
    except Exception as e:
        print(e)
    return {"data": data}
```

```python
tweet_0_labels = get_file("tweet_0_with_labels.json")
```

```python
data=tweet_0_labels["data"]
```

```python
bot_cnt = 0
bot_names = []
for key, datum in data.items():
    if datum["label"] == "bot": 
        bot_names.append(key)
        bot_cnt += 1
```
